```python
import matplotlib, matplotlib.pyplot
matplotlib.rcParams.update({'font.size':20, 
                            'font.family':'sans-serif', 
                            'xtick.labelsize':16, 
                            'ytick.labelsize':16, 
                            'figure.figsize':(16*(2/3), 9*(2/3)), 
                            'axes.labelsize':20
                           })
```

```python
import pandas, numpy
```

```python
input_file = '/Users/adrian/research/egilsstadir/results/deseq2/DESeq2_TPM_values.tsv'
```

```python
df = pandas.read_csv(input_file, sep='\t', index_col=0)
df
```

```python
my_colors = ['gold', 'skyblue', 'tab:red', 'black']
```

```python
# ATG7
target_gene = 'ENSG00000197548' # ATG7
df.loc[target_gene, ]
```

```python
for group in range(4):
    initial = group * 3
    final = initial + 3
    tpm = df.loc[target_gene, ].iloc[initial:final].values
    print(tpm)
    
    matplotlib.pyplot.semilogy(numpy.repeat(group, len(tpm)), tpm, 'o', color=my_colors[group], alpha=3/3, mec='none', markersize=16, base=2)
    matplotlib.pyplot.boxplot(tpm, positions=[group], showcaps=False, medianprops=dict(color='black'))

matplotlib.pyplot.ylim([2**1, 2**5])
yticks_locations = [2, 4, 8, 16, 32]
matplotlib.pyplot.yticks(yticks_locations, yticks_locations)
matplotlib.pyplot.xticks([0, 1, 2, 3], ['KO|+', 'KO|WT', 'KO|KO', 'WT|WT'])
matplotlib.pyplot.ylabel('Expression [TPM]')
matplotlib.pyplot.xlabel('Conditions')
matplotlib.pyplot.grid(ls=':')
matplotlib.pyplot.tight_layout()
matplotlib.pyplot.show()
```
